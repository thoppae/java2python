```python
import pandas as pd

# Example: Load your preprocessed data
train_df = pd.read_csv('train.csv', low_memory=False)

X_train = train_df.drop(['Sales', 'Date'], axis=1)

y_train = train_df['Sales']
```

```python
from sklearn.ensemble import RandomForestRegressor

# Set your model parameters (customize as needed)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)  # Fit model to training data
```

```python
import joblib

# Save your model
joblib.dump(model, 'sales_forecast_model.joblib')
```
